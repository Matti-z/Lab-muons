```python
import pandas as pd
import plotly
import plotly.express as px
import plotly.graph_objects as go
from pathlib import Path
import numpy as np
```

# vecchia analisi

```python
#prendo path per tutti uguale
DATA_FILE = Path('Data') / 'V_settings2.txt'
if not DATA_FILE.exists():
    for p in Path.cwd().parents:
        candidate = p / 'Data' / 'V_settings2.txt'
        if candidate.exists():
            DATA_FILE = candidate
            break
df_variap = pd.read_csv(DATA_FILE, sep = r'\s+', skiprows = 21, header = None, engine='python')
df_variap.columns = ['p_t', 'p_c', 'g_t', 'g_c', 'coincidenze', 'tempo_misura_min', 'thr_p', 'thr_g']
```

```python
DATA_FILE = Path('Data') / 'V_settings3.txt'
if not DATA_FILE.exists():
    for p in Path.cwd().parents:
        candidate = p / 'Data' / 'V_settings3.txt'
        if candidate.exists():
            DATA_FILE = candidate
            break
df_variag = pd.read_csv(DATA_FILE, sep = r'\s+', skiprows = 0, engine='python')
print (df_variag)
```

```python
df_22 = df_variap[df_variap['thr_g'] == 22]
df_35 = df_variap[df_variap['thr_g'] == 35]
df_45 = df_variap[df_variap['thr_g'] == 45]
df_70 = df_variap[df_variap['thr_g'] == 70]
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_22['thr_p'], y=df_22['p_c'], mode='markers', name='thr_g = 22'))
fig.add_trace(go.Scatter(x=df_35['thr_p'], y=df_35['p_c'], mode='markers', name='thr_g = 35'))
fig.add_trace(go.Scatter(x=df_22['thr_p'], y=df_22['coincidenze'], mode = 'markers', name='coinc thr_g = 22'))
fig.add_trace(go.Scatter(x=df_35['thr_p'], y=df_35['coincidenze'], mode = 'markers', name='coinc thr_g = 35'))
fig.add_trace(go.Scatter(x=df_45['thr_p'], y=df_45['p_c'], mode='markers', name='thr_g = 45'))
fig.add_trace(go.Scatter(x=df_45['thr_p'], y=df_45['coincidenze'], mode = 'markers', name='coinc thr_g = 45'))
fig.add_trace(go.Scatter(x=df_70['thr_p'], y=df_70['p_c'], mode='markers', name='thr_g = 70', marker =dict(color = '#1A73E8')))
fig.update_layout(title= 'tensione vs thr', xaxis_title='thr_p', yaxis_title='p_c')
fig.show()
```

```python
df_45p = df_variag[df_variag['thr_p'] == 83]
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_45p['thr_g'], y=df_45p['g_c'], mode='markers', name='thr_g = 45'))
fig.add_trace(go.Scatter(x=df_45['thr_p'], y=df_45['p_c'], mode='markers', name='thr_g = 45'))
```
