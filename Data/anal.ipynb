```python
import pandas as pd
import plotly
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('V_settings2.txt', sep = r'\s+', skiprows = 20, header = None, engine='python')
df.columns = ['p_t', 'p_c', 'g_t', 'g_c', 'coincidenze', 'tempo_misura_min', 'thr_p', 'thr_g']
print (df)
```

```python
df_22 = df[df['thr_g'] == 22]
df_35 = df[df['thr_g'] == 35]
df_45 = df[df['thr_g'] == 45]
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_22['thr_p'], y=df_22['p_c'], mode='markers', name='thr_g = 22'))
fig.add_trace(go.Scatter(x=df_35['thr_p'], y=df_35['p_c'], mode='markers', name='thr_g = 35'))
fig.add_trace(go.Scatter(x=df_22['thr_p'], y=df_22['coincidenze'], mode = 'markers', name='coinc thr_g = 22'))
fig.add_trace(go.Scatter(x=df_35['thr_p'], y=df_35['coincidenze'], mode = 'markers', name='coinc thr_g = 35'))
fig.add_trace(go.Scatter(x=df_45['thr_p'], y=df_45['p_c'], mode='markers', name='thr_g = 45'))
fig.add_trace(go.Scatter(x=df_45['thr_p'], y=df_45['coincidenze'], mode = 'markers', name='coinc thr_g = 45'))
fig.update_layout(title= 'tensione vs thr', xaxis_title='thr_p', yaxis_title='p_c')
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['thr_p'], y=df_22['coincidenze']/df_22['p_c'], mode='markers', name='coinc 22'))
fig.add_trace(go.Scatter(x=df['thr_p'], y=df_35['coincidenze']/df_35['p_c'], mode='markers', name='coin 35'))
fig.add_trace(go.Scatter(x=df['thr_p'], y=df_45['coincidenze']/df_45['p_c'], mode='markers', name='coinc 45'))
```
