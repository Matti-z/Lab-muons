```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from pathlib import Path
import numpy as np
```

```python
#prendo path per tutti uguale
DATA_FILE = Path('Data') / 'ordine_pmg.csv'
if not DATA_FILE.exists():
    for p in Path.cwd().parents:
        candidate = p / 'Data' / 'ordine_pmg.csv'
        if candidate.exists():
            DATA_FILE = candidate
            break
df_variam = pd.read_csv(DATA_FILE, sep = r',', header = 0, engine='python') #skiprow=21
df_1200 = df_variam[df_variam['m_tensione'] == 1200]
df_1240 = df_variam[df_variam['m_tensione'] == 1240]

display(df_1240)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df_1200['m_thr'], 
    y=df_1200['mg_doppia'],
    mode='markers',
    name='mg_doppia',
    error_y=dict(type='data', array=np.sqrt(df_1200['mg_doppia'])),  visible = True))
fig.add_trace(go.Scatter(
    x=df_1200['m_thr'], 
    y=df_1200['pg_doppia'],
    mode='markers',
    name='pg_doppia',
    error_y=dict(type='data', array=np.sqrt(df_1200['pg_doppia'])),  visible = True))
fig.add_trace(go.Scatter(
    x = df_1200['m_thr'],
    y = df_1200['pmg_tripla'],
    mode='markers',
    name='pmg_tripla',
    error_y=dict(type='data', array=np.sqrt(df_1200['pmg_tripla'])), visible = True
))
fig.add_trace(go.Scatter(
    x = df_1200['m_thr'],
    y = df_1200['p_singola'],
    mode='markers',
    name='p_singola',
    error_y=dict(type='data', array=np.sqrt(df_1200['p_singola'])),  visible = True))
fig.add_trace(go.Scatter(
    x = df_1200['m_thr'],
    y = df_1200['pmg_tripla'] / df_1200['pg_doppia'],
    mode='markers',
    name='pmg_tripla/pg_doppia',
    error_y=dict(type='data', array=np.sqrt( ( df_1200['pmg_tripla']/ (df_1200['pg_doppia'])**2 ) + ( (df_1200['pmg_tripla'])**2/ (df_1200['pg_doppia'])**3 ) )),  visible = True))
fig.add_trace(go.Scatter(
    x = df_1240['m_thr'],
    y = df_1240['pmg_tripla'] / df_1240['pg_doppia'],
    mode='markers',
    name='pmg_tripla/pg_doppia 1240',
    error_y=dict(type='data', array=np.sqrt(( df_1240['pmg_tripla']/ (df_1240['pg_doppia'])**2 ) + ( (df_1240['pmg_tripla'])**2/ (df_1240['pg_doppia'])**3 ))),  visible = True))
fig.update_xaxes(title_text='m_thr')
fig.update_yaxes(title_text='Conteggi doppia PMG')
fig.show()
```
